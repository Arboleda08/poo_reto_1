```python
def palindromo(palabra):
  left = 0
  right = len(palabra) - 1
  while left < right:
    if palabra[left] != palabra[right]:
      print(f"La palabra {palabra} no es palindromo")
    else:
      left += 1
      right -= 1
      print(f"La palabra {palabra} es palindromo")

```

```python
palindromo("ana")
```
