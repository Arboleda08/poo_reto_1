```python
def anagrama(lista):
  lista_2 = []
  lista_final = []

  for palabra in lista:
    letras = []
    for letra in palabra:
      letras.append(letra)
    letras.sort()
    lista_2.append(letras)

  for i in range(len(lista)):
    for j in range(len(lista)):
      if i != j and lista_2[i] == lista_2[j]:
        lista_final.append(lista[i])
  return lista_final
```

```python
anagrama(["amor", "roma", "perro"])
```
