```python
def mayor_suma(lista):
  lista_final = []
  for i in range(len(lista)-1):
    suma = lista[i] + lista[i+1]
    lista_final.append(suma)
  mayor = max(lista_final)
  return mayor
```

```python
mayor_suma(lista = [8,20,2,10,40,5])
```
