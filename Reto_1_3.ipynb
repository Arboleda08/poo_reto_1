```python
def lista(x,y):
  x = x
  y = y
  Lista_primos = []
  for n in range(x,y):
    primo = True
    for i in range(2,n):
      if n%i == 0:
        primo = False

    if primo:
      Lista_primos.append(n)
  return Lista_primos
```

```python
lista(5,20)
```
