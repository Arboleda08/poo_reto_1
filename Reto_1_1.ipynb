```python
def calculadora_simple(x, y, operacion):
  if operacion == "+":
    return x + y
  elif operacion == "-":
    return x - y
  elif operacion == "/":
    return x / y
  elif operacion == "*":
    return x * 7
  else:
    print("la operacion que usted ingreso no esta disponible")
```
